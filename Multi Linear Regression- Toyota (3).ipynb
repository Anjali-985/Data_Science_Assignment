## Model Diagnostics

Detecting Outliers/Influencers

## Cook's Distance Method

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
from statsmodels.graphics.regressionplots import influence_plot
```

```python
#Plot the influencers values using stem plot

fig = plt.subplots(figsize=(22,9))
plt.stem(np.arange(len(toyotaRename)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#Showing the index and value of influencer for which  - -> c > 0.5
# arrgmax = Returns the indices of max value along an axis
print("The index of Outlier :", np.argmax(c), "The value of the outlier is :", np.max(c))
```

## Leverage value using Influence Points

```python
#Points beyond leverage cutoff values are influencers
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

## Model Deletion Diagnostics & Final Model

```python
model.summary()
```

```python
# Diagnostics and Model Deletion method by using np.max ONLY
while np.max(c)>0.5 :
    model=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+QT+Weight',data=Toyota1).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    Toyota1 = Toyota1.drop(Toyota1.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    Toyota1
else:
    final_model=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+QT+Weight',data=Toyota1).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```

```python
# Diagnostics and Model Deletion method by r squared & using np.max ONLY

while model.rsquared < 0.90:
    for c in [np.max(c)>0.5]:
        model=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+QT+Weight',data=Toyota1).fit()
        (c,_)=model.get_influence().cooks_distance
        c
        np.argmax(c) , np.max(c)
        Toyota1=Toyota1.drop(Toyota1.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        Toyota1
    else:
        final_model1=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+QT+Weight',data=Toyota1).fit()
        final_model1.rsquared , final_model1.aic
        print("Thus model accuracy is improved to",final_model1.rsquared)
```
